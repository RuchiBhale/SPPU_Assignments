```python
text="In a bustling metropolis, the city lights flickered like a galaxy of stars against the night sky. Cars streamed through the streets like ants on a mission, their headlights cutting through the darkness. Pedestrians hurried along the sidewalks, their footsteps creating a rhythmic beat against the pavement. The aroma of street food wafted through the air, tempting passersby with its tantalizing scent. Amidst the urban chaos, a sense of vibrant energy pulsed through the city, a testament to its ceaseless vitality."
```

```python
from nltk.tokenize import word_tokenize
word_tokenize(text)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
text_tokens = word_tokenize(text)
tokens_without_sw= [word for word in text_tokens if not word in stop_words]

print(tokens_without_sw)
```

```python
print(stop_words)
```

```python
parsed_text = word_tokenize(text)
# Initialize stemmer.
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer('english')

# Stem each word.
[(word, stemmer.stem(word)) for i, word in enumerate(parsed_text)
 if word.lower() != stemmer.stem(parsed_text[i])]
```

```python
# import these modules
from nltk.stem import WordNetLemmatizer
print(parsed_text)

lemmatizer = WordNetLemmatizer()
[(word, lemmatizer.lemmatize(word)) for i, word in enumerate(parsed_text)
 if word.lower() != lemmatizer.lemmatize(parsed_text[i]).lower()]

```

```python
#Part of Speech tagging(POS)
pos_tags=nltk.pos_tag(parsed_text)
print(pos_tags)
```

```python

# import required module
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# create object
tfidf = TfidfVectorizer()

# get tf-df values
result = tfidf.fit_transform(parsed_text)
```

```python
print(result)
```

```python

# get idf values
print('\nidf values:')
for ele1, ele2 in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(ele1, ':', ele2)
```
