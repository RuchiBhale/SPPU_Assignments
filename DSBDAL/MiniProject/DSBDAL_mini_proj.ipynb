```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('covid_vaccine_statewise.csv')
```

```python
df[212:250]
```

```python
df['State'].unique()
```

```python
df = df.drop(df[df['State'] == 'India'].index)
```

```python
df['State'].unique()
```

```python
df.index=df.index-212
```

```python
df.head()
```

```python
df=df.drop(['Male(Individuals Vaccinated)','Female(Individuals Vaccinated)','Transgender(Individuals Vaccinated)'],axis=1)
```

```python
df=df.drop(['18-44 Years (Doses Administered)',	'45-60 Years (Doses Administered)',	'60+ Years (Doses Administered)','18-44 Years(Individuals Vaccinated)'	,'45-60 Years(Individuals Vaccinated)','60+ Years(Individuals Vaccinated)'],axis=1)
```

```python
cols_keep=['Updated On','State','Total Doses Administered','Sessions',' Sites ','First Dose Administered','Second Dose Administered','Male (Doses Administered)','Female (Doses Administered)']
df=df[cols_keep]
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.head()
```

No of persons state wise vaccinated for first dose

```python
state_sum = df.groupby('State')['First Dose Administered'].sum()

print("State\t\t\t\t\tCount of 1st dose")
for state, total_dose in state_sum.items():
    print(state, "\t\t\t\t\t", int(total_dose))
```

Number of persons state wise vaccinated for second dose

```python
state_sum = df.groupby('State')['Second Dose Administered'].sum()

print("State\t\t\t\t\tCount of 2nd dose")
for state, total_dose in state_sum.items():
    print(state, "\t\t\t\t\t", int(total_dose))
```

Number of Males Vaccinated
Number of Females Vaccinated

```python
males=df['Male (Doses Administered)'].sum()
females=df['Female (Doses Administered)'].sum()
print("Males vaccinated",int(males))
print("females vaccinated",int(females))
```

```python
total_male_doses = df['Male (Doses Administered)'].sum()
total_female_doses = df['Female (Doses Administered)'].sum()

# Pie chart
labels = ['Male', 'Female']
sizes = [total_male_doses, total_female_doses]
colors = ['pink','skyblue']
explode = (0, 0.1)

plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Doses Administered between Males and Females')
plt.axis('equal')
plt.show()
```

```python
state_sum = df.groupby('State')['Total Doses Administered'].sum()

# Plotting
plt.figure(figsize=(15, 8))
state_sum.plot(kind='bar', color='skyblue')
plt.title('Sum of Total Doses Administered by State')
plt.xlabel('State')
plt.ylabel('Sum of Total Doses Administered')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
